```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
file_path = '/Users/liya/Desktop/cattle_label/TopViewDairyCattle-HeatmapImages-YL'
code_list = []
for root,dirs,files in os.walk(file_path):
    for filename in files:
        if filename[-4:] == '.csv':
            code_list.append(filename[:-4])
code_list.sort()
print(code_list)
np.size(code_list) 
```

```python
path = './cattlepic'
if not os.path.exists(path):
    os.makedirs(path)
```

```python
for i in range(81):
    csv_data = pd.read_csv(code_list[i]+'.csv',header = None)
    pic = sns.heatmap(csv_data,cbar=False)
    s1 = pic.get_figure()
    s1.savefig(os.path.join(path,code_list[i]+'.png'))
```
